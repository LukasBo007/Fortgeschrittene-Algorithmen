```python
import csv
import matplotlib.pyplot as plt
import math
import numpy as np
from bitarray import bitarray
import random

#ASCII CODE erstellen
def keyWert(inputwert):
    return sum(ord(buchstabe) for buchstabe in inputwert)

#Erste Hashfunktion( Division)
#Dublikate werden mit einer Primzahl multipliziert und rekursiv als neuen Key in die Funktion eingefügt.
def divisionHash(input_key: int, hashsize, maximaleRekursion=1031, startRekursion=0):
    hashWert = input_key % hashsize
    if hashWert in hashWerts_division and startRekursion < maximaleRekursion:
        return divisionHash(hashWert * 31, hashsize, maximaleRekursion, startRekursion + 1)
    return hashWert

#Zweite Hash Funktion( Multiplikation)
#Dublikate werden mit einer Primzahl multipliziert und rekursiv als neuen Key in die Funktion eingefügt.
def multiplikationHash(input_key: int, hashsize, maximaleRekursion=1031, startRekursion=0):
    A = (math.sqrt(5) - 1) / 2 #Goldener Schnitt Phi für bessere Ergebnisse
    hashWert = int(hashsize * (input_key * A % 1))
    if hashWert in hashWerts_multiplication and startRekursion < maximaleRekursion:
        return multiplikationHash(hashWert * 7, hashsize, maximaleRekursion, startRekursion + 1)
    return hashWert

#Bloomfilter definieren
class BloomFilter:
    def __init__(self, n, p, hashsize):
        self.m = math.ceil((n * math.log(p)) / math.log(1 / pow(2, math.log(2))))
        self.k = 2  
        self.bit_array = bitarray(self.m)
        self.bit_array.setall(0)
        self.hashsize = hashsize

    def insert(self, element):
        key = keyWert(element)
        indices =  [divisionHash(key, self.hashsize) % self.m, multiplikationHash(key, self.hashsize) % self.m]
        for index in indices:
            self.bit_array[index] = 1

    def search(self, element):
        key = keyWert(element)
        indices =  [divisionHash(key, self.hashsize) % self.m, multiplikationHash(key, self.hashsize) % self.m]
        return all(self.bit_array[index] for index in indices)

# CSV-Datei lesen
def read_csv_file(file_path):
    #Encoding "iso"" passt gut für spezielle Buchstaben
    with open(file_path, 'r', encoding='ISO-8859-1') as file:
        csv_reader = csv.reader(file)
        return [str(row[0]) for row in csv_reader]

#Zielpfad definieren und alle Wörter eines Sprachpakets benutzen
file_path = "/Users/furkan/Downloads/archive/de_DE.csv"
all_words = read_csv_file(file_path)

# Zufällig 10.000 Wörter zum Testen auswählen
test_elements = random.sample(all_words, 10000)

# Initialisieren des Bloom-Filters
n = 10000
p = 0.05
bloom_filter = BloomFilter(n, p, hashsize)

#Liste für die Outputwerte der Hashfunktionen
hashWerts_division = {}
hashWerts_multiplication = {}

# Einfügen von Elementen
for word in all_words:
    key = keyWert(word)
    hashwert_division = divisionHash(key, hashsize)
    hashwert_multiplication = multiplikationHash(key, hashsize)
    hashWerts_division[hashwert_division] = hashWerts_division.get(hashwert_division, 0) + 1
    hashWerts_multiplication[hashwert_multiplication] = hashWerts_multiplication.get(hashwert_multiplication, 0) + 1

# Anzahl der belegten Plätze für beide Funktionen
unique_places_division = len(hashWerts_division)
unique_places_multiplication = len(hashWerts_multiplication)

print(f"Anzahl der belegten Plätze mit divisionHash: {unique_places_division}")
print(f"Anzahl der belegten Plätze mit multiplikationHash: {unique_places_multiplication}")

# Testen des Bloom-Filters
false_positives = 0
for element in test_elements:
    if bloom_filter.search(element):
        false_positives += 1

print(f"Größe des Bloom-Filters (m): {bloom_filter.m}")
print(f"Anzahl der Hash-Funktionen (k): {bloom_filter.k}")
print(f"Anzahl der falsch positiven Ergebnisse: {false_positives}")

#Histogramm
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.bar(hashWerts_division.keys(), hashWerts_division.values(), color='skyblue')
plt.title('Histogramm für divisionHash')
plt.ylim(0, 3)

plt.subplot(1, 2, 2)
plt.bar(hashWerts_multiplication.keys(), hashWerts_multiplication.values(), color='lightcoral')
plt.title('Histogramm für multiplikationHash')
plt.ylim(0, 3)

plt.tight_layout()
plt.show()
```
